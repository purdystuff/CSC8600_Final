```python

from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.feature_extraction.text import TfidfVectorizer

movies_df = pd.read_csv("Datasets/clean_movies.csv")
# cast_df = pd.read_csv("Datasets/clean_cast.csv")
# crew_df = pd.read_csv("Datasets/clean_crew.csv")
# keyword_df = pd.read_csv("Datasets/keywords.csv")
# keyword_df = pd.read_csv("Datasets/clean_keyword.csv")
# genres_df = pd.read_csv("Datasets/clean_genres.csv")
# production_df = pd.read_csv("Datasets/clean_production.csv")

movies_df['movie_text'] = movies_df['movie_text'].fillna('')

X = movies_df['movie_text']
```

```python
from sklearn.cluster import KMeans

model = TfidfVectorizer(stop_words='english', max_features=1000)
X = model.fit_transform(movies_df['movie_text'])

kmeans = KMeans(n_clusters=10, random_state=42)
movies_df['cluster'] = kmeans.fit_predict(X)


print(movies_df[movies_df['cluster'] == 3]['title'].head(100))
```

```python
print(movies_df['cluster'].value_counts())
```

```python
cluster_counts = movies_df['cluster'].value_counts()

terms = model.get_feature_names_out()
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]

print("Top Clusters by Size:")
print("-" * 30)

for cluster_id, count in cluster_counts.items():
    top_terms = [terms[ind] for ind in order_centroids[cluster_id, :5]]
    
    print(f"Cluster {cluster_id} ({count} movies):")
    print(f" {', '.join(top_terms)}\n")
```
