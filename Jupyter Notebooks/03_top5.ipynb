```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

movies_df = pd.read_csv("Datasets/clean_movies.csv")
cast_df = pd.read_csv("Datasets/clean_cast.csv")

movies_df['profit'] = movies_df['revenue'] - movies_df['budget']
movies_df['not_flop'] = (movies_df['profit'] > 0)
```

```python
cast_overlap = pd.merge(movies_df, cast_df, left_on='id', right_on='movie_id')
```

```python
cast_overlap
```

```python
jackson_movies = cast_overlap[cast_overlap['actor_name'].str.contains("Samuel L. Jackson")]
paul_movies = cast_overlap[cast_overlap['actor_name'].str.contains("Paul Giamatti")]

```

```python
jackson_movies[['original_title', 'actor_name', 'profit', 'not_flop']].sort_values('profit',ascending=False)
```

```python
paul_movies[['original_title', 'actor_name', 'profit']].sort_values('profit',ascending=False)
```

```python
print("$" , jackson_movies['profit'].sum())
print("$" , paul_movies['profit'].sum())
```

```python
print(jackson_movies['not_flop'].value_counts())
print(paul_movies['not_flop'].value_counts())
```
